```python
fetch/show information from table
select statment

1- show all table

code          name     salary   age

1             moh       5000     NULL
2             ayman     NULL     NULL

import MySQLdb
con = MySQLdb.connect("localhost", "root", "", "first_db")
cur = con.cursor()

SQL_commands = "select * from company"
try:
    cur.execute(SQL_commands)
    var = cur.fetchall()
    print(var)
    for column in var:
        code = column[0]
        name = column[1]
        salary = column[2]
        age = column[3]
        print(code, name, salary, age)
except:
    print("not executed")
con.close

result:
((1L, 'moh', 5000L, None), (2L, 'ayman', None, None))
---------------------------------------------------------
(1L, 'moh', 5000L, None)
(2L, 'ayman', None, None)
```
